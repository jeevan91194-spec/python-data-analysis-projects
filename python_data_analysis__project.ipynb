##Sales Data Analysis

###Data analysis,
###Visualization,
###Business insights

```python
import pandas as pd
import matplotlib.pyplot as plt

# load dataset
df = pd.read_csv("/content/Superstore Sales Dataset.csv")

# total revenue
total_revenue = df["Sales"].sum()
print("Total Revenue:", total_revenue)

# sales by region
sales_region = df.groupby("Region")["Sales"].sum()
print(sales_region)

# visualization
sales_region.plot(kind="bar")
plt.title("Sales by Region")
plt.show()

```

###Customer Segmentation
high value customers

```python
customer_revenue = df.groupby("Customer ID")["Sales"].sum()
top_customers = customer_revenue.sort_values(ascending=False).head(10)
print(top_customers)
```

#Sales Trend Forecast
Time series analysis

```python
df["Order Date"] = pd.to_datetime(df["Order Date"], dayfirst=True)

monthly_sales = df.groupby(df["Order Date"].dt.month)["Sales"].sum()

monthly_sales.plot()
plt.title("Monthly Sales Trend")
plt.show()

```
